```python
#imports
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.activations import tanh
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, LeakyReLU, MaxPooling2D, Flatten, Dense, Activation
from tensorflow.keras.optimizers import SGD, Adam

#loading the Cirfar-10 dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
#preparing the data
y_train = y_train.flatten()
y_test = y_test.flatten()
#classes of the images
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
#reshaping
x_train= (x_train.reshape(x_train.shape[0], x_train.shape[1], x_train.shape[2], 3))/ 255.0
x_test = (x_test.reshape(x_test.shape[0], x_test.shape[1], x_test.shape[2], 3))/ 255.0
#converting label to one-hot encoding
y_train = tf.one_hot(y_train.astype(np.int32), depth=10)
y_test = tf.one_hot(y_test.astype(np.int32), depth=10)

# plotting image
plt.imshow(x_train[10])
print(y_train[10])

# CNN model with LeakyRelu activation function & SGD optimizer
model_1 = Sequential([
  Conv2D(32, (3,3), activation='linear',padding='same', input_shape=x_train.shape[1:]),
  LeakyReLU(alpha=0.1),
  MaxPooling2D(pool_size=(2,2)),
  Conv2D(64, (3,3), activation='linear', padding='same'),
  LeakyReLU(alpha=0.1),
  MaxPooling2D(pool_size=(2,2)),
  Flatten(),
  Dense(10, activation='softmax')
])

# CNN model with LeakyRelu activation function & SGD optimizer
model_1 = Sequential([
  Conv2D(32, (3,3), activation='linear',padding='same', input_shape=x_train.shape[1:]),
  LeakyReLU(alpha=0.1),
  MaxPooling2D(pool_size=(2,2)),
  Conv2D(64, (3,3), activation='linear', padding='same'),
  LeakyReLU(alpha=0.1),
  MaxPooling2D(pool_size=(2,2)),
  Flatten(),
  Dense(10, activation='softmax')
])

# CNN model with LeakyRelu activation function & SGD optimizer
model_3 = Sequential([
  Conv2D(32, (3,3), activation='linear', padding='same', input_shape=x_train.shape[1:]),
  Activation(tanh),
  MaxPooling2D(pool_size=(2,2)),
  Conv2D(64, (3,3), activation='linear', padding='same'),
  Activation(tanh),
  MaxPooling2D(pool_size=(2,2)),
  Flatten(),
  Dense(10, activation='softmax')
])


def CTET(model,optimizer):
    # Compile the model
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    # Train the model
    history = model.fit(x_train, y_train, epochs=30, batch_size=32)
    # Evaluate training
    fig, ax = plt.subplots(2,1)
    ax[0].plot(history.history['loss'], color='b', label="Training Loss")
    legend = ax[0].legend(loc='best', shadow=True)
    ax[1].plot(history.history['accuracy'], color='b', label="Training Accuracy")
    legend = ax[1].legend(loc='best', shadow=True)
    # Test the model
    test_loss, test_acc = model.evaluate(x_test, y_test)
    plt.show()
    print('Test accuracy:', test_acc)

```

Run the experiment and report accuracy on test set with:

- CNN network for classifying images, using LeakyReLU as the activation function, Adam as the optimizer and 0.0001 as the learning rate
- CNN network for classifying images, using Tanh as the activation function, SGD as the optimizer and 0.0001 as the learning rate


```python
# CNN model with LeakyRelu activation function & SGD optimizer
CTET(model_1,SGD(learning_rate=0.0001))

# CNN model with LeakyRelu activation function & Adam optimizer
CTET(model_1,Adam(learning_rate=0.0001))

# CNN model with Tahn activation function & SGD optimizer
CTET(model_3,SGD(learning_rate=0.0001))

```
